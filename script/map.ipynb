```python
import random

def generate_divided_grid():
    grid_size = 8  # グリッドのサイズ
    line_count = 4  # 線の数

    # グリッドを初期化して全体を壁(#)で埋める
    grid = [['.' for _ in range(grid_size)] for _ in range(grid_size)]

    # 各線ごとにランダムに穴を開ける
    for _ in range(1):
        # 線をランダムに選ぶ
        line = random.choice(['horizontal', 'vertical'])

        if line == 'horizontal':
            # 水平線をランダムな行に描く
            row = random.randint(1, grid_size - 2)
            for i in range(grid_size):
                grid[row][i] = "#"
            hole = random.randint(1, grid_size - 2)
            grid[row][hole] = '.'
        elif line == 'vertical':
            # 垂直線をランダムな列に描く
            col = random.randint(1, grid_size - 2)
            hole = random.randint(1, grid_size - 2)
            grid[hole][col] = '.'

    return grid

def print_grid(grid):
    for row in grid:
        print(' '.join(row))

grid = generate_divided_grid()
print_grid(grid)
```
